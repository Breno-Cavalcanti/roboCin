```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set()
%matplotlib inline
```

```python
data = pd.read_csv('test_std_with_vanishing2.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data['real_value'] = data['real_value'].replace(0, np.nan)
```

```python
data.shape
```

```python
data.head()
```

```python
data.reset_index(inplace= True, drop = True)
```

```python
data.drop(0, inplace= True)
data.head(10)
```

```python
#data.to_csv('data2_with_nan.csv')
```

```python
plt.figure(figsize= (14, 7))
#plt.axis([100,200, -5000, -4500])
plt.title('Comparação Real x Predição')
plt.plot(data['real_value'], marker = '*' ,label = 'Valores reais')
plt.plot(data['predicted_value'], label =  'Previsão do filtro')
plt.legend()
plt.savefig('real_std_perfect.png', format='png')
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(data['real_value'], data['predicted_value'])
```

```python
plt.figure(figsize= (14, 7))
plt.axis([100,200, -3500, -2500])
plt.title('Comparação Real x Predição')
plt.plot()
plt.plot(data['real_value'], label = 'Valores reais')
plt.plot(data['predicted_value'], label =  'Previsão do filtro')
plt.legend()
plt.savefig('real_std_zoom_perfect.png', format='png')
```
