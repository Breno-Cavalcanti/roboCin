```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set()
%matplotlib inline
```

```python
data = pd.read_csv('./data/test_std_with_correction_int.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data['real_value'] = data['real_value'].replace(0, np.nan)
```

```python
plt.figure(figsize= (14, 7))
#plt.axis([145,160, -4760, -4600])
plt.title('Comparação Real x Predição')
plt.plot()
plt.plot(data['real_value'], marker = '*' , label = 'Valores reais')
plt.plot(data['predicted_value'], label =  'Previsão do filtro')
plt.savefig('./imgs/real_std_without_zoom.png', format='png')
plt.legend()
```

```python
plt.figure(figsize= (14, 7))
plt.axis([100,200, -5000, -4600])
plt.title('Comparação Real x Predição')
plt.plot()
plt.plot(data['real_value'], marker = '*' , label = 'Valores reais')
plt.plot(data['predicted_value'], label =  'Previsão do filtro')
plt.legend()
plt.savefig('./imgs/real_std_with_zoom.png', format='png')
plt.show()
```
