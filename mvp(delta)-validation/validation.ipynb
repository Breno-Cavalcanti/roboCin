```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set()
%matplotlib inline
```

```python
data = pd.read_csv('validation_mvp.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data['real_values'] = data['real_values'].replace(0, np.nan)
```

```python
data.head()
```

```python
plt.figure(figsize= (14, 7))
plt.axis([100,200, -5000, -4000])
plt.title('Comparação Real x Predição')
plt.plot(data['real_values'], marker = '*' ,label = 'Valores reais')
plt.plot(data['predicted_values'], label =  'Previsão do filtro')
plt.legend()
plt.savefig('zoom_mvp.png', format='png')
plt.show()
```
