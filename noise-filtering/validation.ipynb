```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
data = pd.read_csv('positions_with_noise.csv')
```

```python
data.head()
```

```python
plt.figure(figsize= (14, 7))
plt.title('Comparação Ruído X Friltragem')
plt.plot(data['position_noise'], marker = '*' ,label = 'Valores com ruído')
plt.plot(data['position_filtred'],label =  'Valores filtrados')
plt.legend()
plt.savefig('noise-filtering.png', format='png')
plt.show()
```

```python
plt.figure(figsize= (14, 7))
plt.title('Comparação Ruído X Friltragem')
plt.axis([800,850, 0, 1000])
plt.plot(data['position_noise'] ,label = 'Valores com ruído')
plt.plot(data['position_filtred'], marker = 'o',label =  'Valores filtrados')
plt.legend()
plt.savefig('zoom-incress_vel.png', format='png')
plt.show()
```

```python
plt.figure(figsize= (14, 7))
plt.title('Comparação Ruído X Friltragem')
plt.axis([740,800, 0,400 ])
plt.plot(data['position_noise'] ,label = 'Valores com ruído')
plt.plot(data['position_filtred'], marker = 'o',label =  'Valores filtrados')
plt.legend()
plt.savefig('zoom-decrease-vel.png', format='png')
plt.show()
```
